```python
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as py
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns

COLOR = 'white'
plt.rcParams['text.color'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
```

```python
champ = pd.read_csv('wc_champions.csv')
champ.head()
```

```python
del champ['Unnamed: 0']
champ.head()
```

```python
champ['winrate'] = list(map(lambda x: x[:-1], champ['winrate_total'].values))
# 퍼센트 제거하는 코드 (퍼센트가 있어서 계산이 불가능해서)
# 한번만 할 것 왜냐면 뒤에 문자 및 숫자를 삭제를 하는 코드라 계속하면 데이터가 사라짐
champ['winrate'] = champ['winrate'].astype('float64') * 0.01
champ['winrate']
champ.head()
```

```python
# 5회이상 선택된 챔피언 중 가장 높은 승률
champs = champ[(champ['sum_total'] > 5)]
champs['winrate'].max()
```

```python
highest_win = champs[champs['winrate'] == champs['winrate'].values.max()]
highest_win
print('The most highest winrate champions that more than 5 times selected were' , highest_win['champion'][48],'and', highest_win['champion'][49])
```

```python
# 30번 이상 선택에 승률 5할 이상인 챔피언들
champ_30 = champs[(champs['sum_total'] > 30) & (champs['winrate'] > 0.50)]
champ_30['champion'].to_frame()
# plt.title('how many game most champions win')
# plt.show()
```

```python
players = pd.read_csv('wc_players.csv')
players.head()
```

```python
ban = players[['ban1', 'ban2', 'ban3', 'ban4', 'ban5']]
ban['ban4'].value_counts()
```

```python
# 모든 경기에서 밴 당한 총 챔피언들의 수의 합

ban_total = pd.DataFrame([ban['ban1'].value_counts(), ban['ban2'].value_counts(), ban['ban3'].value_counts(), ban['ban4'].value_counts(), ban['ban5'].value_counts()]).T

ban_frame = ban_total.sum(axis=1).to_frame()
ban_frame = ban_frame / 5

ban_frame.columns = ['ban_counts']
ban_frame = ban_frame.sort_values(by=['ban_counts'], ascending=False)
ban_frame.head()
```
